```python
import datetime
from functools import reduce
from datetime import date
import json
import csv

c_date = datetime.date.today() 

def process(My_Dict):

    alp_chr = [chr(i) for i in range(97,104)]  # 'a', 'b', 'c', 'd', 'e', 'f', 'g'
    alp_chr= tuple(alp_chr)
    
    alp_filter = [chr(x) for x in [107, 109, 115]]  # 'k','m','s'
    alp_filter = tuple(alp_filter)
    
    chr_filtering = list(filter(lambda x: x[1]["name"][0].lower() not in (alp_filter), My_Dict.items()))
    My_Dict = dict(chr_filtering)

    mapping = list(map(lambda x: "Mr. " + x if str(x).lower().startswith((alp_chr))else "Ms. " + x,[My_Dict[x]['name'] for x in My_Dict]))
        
    age_year = list(map(lambda y: (date.today().year) - y ,[ My_Dict[i]['age'] for i in My_Dict]))
 
    from functools import reduce
    total_sum = reduce(lambda x, y: x + y, (My_Dict[i]['amount_deposited'] for i in My_Dict))
    
    i = 0
    for x in My_Dict:
        My_Dict[x]['name'] = mapping[i]
        My_Dict[x]['year'] = age_year[i]
        My_Dict[x]['amount_deposited'] = total_sum[i]
        i = 1 + i
    return My_Dict
    # print(My_Dict)

# def wrrite_file(final_data):
#     try:
#         with open('Assignment.json', 'w') as file:
#             json.dump(final_data, file, indent=4)
#         return True
#     except Exception as e:
#         return False

    
def main():
    My_Dict = {
'person_1':{'name' : 'Abdul Rafay','age':22, 'Interests':['football','cricket'], 'amount_deposited':
[24000,26000] },'person_2':{'name' : 'Nancy James','age':23, 'Interests':['baseball','cricket'], 'amount_deposited':
[24000,27000] }, 'person_3':{'name' : 'Selena Gomez','age':26, 'Interests':['baseball','table tennis'], 'amount_deposited':
[24000,28000] }}

    data =  process(My_Dict)

    final_data = {
        'Current_Date': c_date.strftime('%d-%b-%Y'),
        'Data': data
    }
    
    print(final_data)

    
#     list = [['20+30', '50', 'pass'],['10+12', '33', 'fail']]
#     with open('names.csv','w') as f:
#         writer = csv.writer(f)
#         writer.writerow(['name', 'Interests', 'amount_deposited','year'])
#          write.writerows(rows)
#         writer.writerows(final_data) 
#         print('The CSV file has been saved')

        
#     header = ['name', 'Interests','age', 'amount_deposited','year']

#     with open('names.csv', 'w') as file:
#     writer = csv.DictWriter(, fieldnames = field_names)
#     writer.writeheader()
#     writer.writerows(cars)

    with open('names.json', 'w') as file:
        json.dump(final_data, file, indent=2, sort_keys=False)
        print('The Json file has been saved')

if __name__ == '__main__':
    main()
```
